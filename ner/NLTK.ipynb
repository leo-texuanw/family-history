# Generate bio format corpus

```python
import pickle
from os.path import join

OBJECT_DIR = "../objects/"
CORPUS_DIR = "../Data/bioc_FH_training/"

ent_files = pickle.load(open(join(OBJECT_DIR, "entities_corrected.pkl"), "rb"))
```

```python
print(ent_files)
```

```python
from collections import defaultdict

doc_ents_dict = defaultdict(dict)
for doc_id, ents in ent_files.items():
    for ent in ents:
        doc_ents_dict[doc_id][ent[2]] = doc_ents_dict.get(doc_id, {}).get(ent[2], []) + ent[3].split(" ")

doc_ents_dict
```

```python
doc_labels = {}

doc_sents = pickle.load(open(join(OBJECT_DIR, "sents.pkl"), "rb"))
for doc_id, sents in doc_sents.items():

#    with open(join(PROCESSED_DIR, doc_id+".bio"), "wb") as bio_file:
        labels = []
        for sent in sents:
            for token in sent.split():
                if token in doc_ents_dict.get(doc_id, {}).get('FamilyMember', []):
#                    bio_file.write(' '.join([token, 'FMS\n']).encode())
                    labels.append((token, 'FMS'))
                elif token in doc_ents_dict.get(doc_id, {}).get('Observation', []):
#                    bio_file.write(' '.join([token, 'OBS\n']).encode())
                    labels.append((token, 'OBS'))
                else:
#                    bio_file.write(' '.join([token, 'O\n']).encode())
                    labels.append((token, 'O'))
        doc_labels[doc_id] = labels

doc_labels
```

# Traning

```python
PROCESSED_DIR = '../Data/processed/'

with open(join(PROCESSED_DIR, 'train_set.txt'), 'r') as f_train_set:
    train_set = ['doc_' + doc_id.strip('\n') for doc_id in f_train_set.readlines()]

with open(join(PROCESSED_DIR, 'test_set.txt'), 'r') as f_test_set:
    test_set = ['doc_' + doc_id.strip('\n') for doc_id in f_test_set]

# from os import listdir
# CORPUS_DIR = '../Data/bioc_FH_training/'
# corpus = [file for file in listdir(CORPUS_DIR) if file.endswith('.pkl')]
```

```python
with open(join(PROCESSED_DIR, 'train_bio.tsv'), 'wb') as f_train_bio:
    for doc_id in train_set:
        for token, label in doc_labels[doc_id]:
            f_train_bio.write('\t'.join([token, label+'\n']).encode())
```

```python
import nltk
from nltk.tag.stanford import StanfordNERTagger

jar = '/home/texuanw/tools/stanford-ner-2018-10-16/stanford-ner.jar'
model = './fh_ner_model.ser.gz'

ner_tagger = StanfordNERTagger(model, jar, encoding='utf8')

for doc_id in test_set:
    with open(join(CORPUS_DIR, doc_id+'.txt'), 'r') as txt:
        words = nltk.word_tokenize(txt.read())
        print(ner_tagger.tag(words))
```
